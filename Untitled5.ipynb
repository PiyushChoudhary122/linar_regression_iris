```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
X=[4,5,10,4,10,8,3,2,7,3]
Y=[23,27,22,30,21,23,27,29,30,25]

data=list(zip(X,Y))
data
```

```python
model = KMeans(n_clusters =2)
model.fit(data)
```

```python
plt.scatter(X,Y, c=model.labels_)
plt.show()
```
